```python
import numpy as np
import pandas as pd
import keras
from keras.utils import np_utils
```

```python
mnist = keras.datasets.mnist.load_data()
(x_Train, y_Train), (x_Test, y_Test) = mnist

print('x_train_image:',x_Train.shape)
print('y_train_label:',y_Train.shape)
print('x_test_image:',x_Test.shape)
print('y_test_label:',y_Test.shape)
```

```python
import matplotlib.pyplot as plt
def plot_image(image):
    fig = plt.gcf()
    fig.set_size_inches(2, 2)
    plt.imshow(image, cmap='binary')
    plt.show()
```

```python
# 由於圖片通常是RGB三個顏色所組成的，假設圖片大小是28*28的彩色圖片，實際上的資料維度就是28*28*3。
# 不過這邊數字的顏色都是單色因此我們轉成28*28*1的資料維度當作未來CNN Model的input
print("x_Train[0].shape:", x_Train[0].shape)
print("y_Train[0]:", y_Train[0])
plot_image(x_Train[0])
```

```python
import matplotlib.pyplot as plt
def plot_images_labels_prediction(images,labels,
                                  prediction,idx,num=10):
    fig = plt.gcf()
    fig.set_size_inches(12, 14)
    if num>25: num=25 
    for i in range(0, num):
        ax=plt.subplot(5,5, 1+i)
        ax.imshow(images[idx], cmap='binary')
        title= "label=" +str(labels[idx])
        if len(prediction)>0:
            title+=",predict="+str(prediction[idx]) 
            
        ax.set_title(title,fontsize=10) 
        ax.set_xticks([]);ax.set_yticks([])        
        idx+=1 
    plt.show()
```

```python
# 訓練集
plot_images_labels_prediction(x_Train,y_Train,[],0,10)
```

```python
# 測試集
plot_images_labels_prediction(x_Test,y_Test,[],0,10)
```

```python
x_Train.shape
```

```python
# 多加一個顏色的維度 
x_Train4D=x_Train.reshape(x_Train.shape[0],28,28,1).astype('float32')
x_Test4D=x_Test.reshape(x_Test.shape[0],28,28,1).astype('float32')
```

```python
print(x_Train4D.shape)
# 需将[28, 28, 1]squeeze为[28, 28]
plot_image(x_Train4D[0].squeeze())
```

```python
# 將數值縮小到0~1 normalization
x_Train4D_normalize = x_Train4D / 255
x_Test4D_normalize = x_Test4D / 255

print(x_Train4D_normalize[0].shape)
plot_image(x_Train4D_normalize[0].squeeze())
```

```python
# 把類別做Onehot encoding
y_TrainOneHot = np_utils.to_categorical(y_Train)
y_TestOneHot = np_utils.to_categorical(y_Test)

# 看來有10種類型的 解答 0 to 9
print(y_TrainOneHot[0])
print(y_TestOneHot[0])
```

```python
y_Train
```

## 建立CNN模型

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
```

```python
import os
save_dir = os.path.join(os.getcwd(), 'saved_models')
model_name = 'keras_mnist_trained_model.h5'
model_path = os.path.join(save_dir, model_name)
```

```python
model = Sequential()
```

```python
#ReLu function
def relu(X):
   return np.maximum(0,X)

print("relu(-1.111)", relu(-1.111))
print("relu(-100)", relu(-100))
print("relu(0.9)", relu(0.9))
print("relu(10)", relu(10))
```

```python
#filter為16, Kernel size為(5,5),Padding為(same), 圖片大小 (28,28,1)
model.add(Conv2D(filters=16,
                 kernel_size=(5,5),
                 padding='same',
                 input_shape=(28,28,1), 
                 activation='relu'))
```

```python
# MaxPooling(取最大值) size為(2,2)
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Conv2D(filters=36,
                 kernel_size=(5,5),
                 padding='same',
                 activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
# Drop掉部分神經元避免overfitting
model.add(Dropout(0.25))
```

```python
# 平坦化
model.add(Flatten())
```

```python
model.add(Dense(128, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(10,activation='softmax'))
```

```python
print(model.summary())
```

## 訓練模型

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
train_history=model.fit(x=x_Train4D_normalize, 
                        y=y_TrainOneHot,validation_split=0.2, 
                        epochs=20, batch_size=300,verbose=2)
```

```python
## Save model and weights
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
model.save(model_path)
print('Saved trained model at %s ' % model_path)


## load model and weights
# from keras.models import load_model
# model = load_model(model_path)
```

```python
prediction = model.predict_classes(x_Test4D_normalize[0:2])
print('test after load: ', prediction)
plot_images_labels_prediction(x_Test[0:2],y_Test[0:2],prediction,0,2)
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_acc,test_acc):
    plt.plot(train_history.history[train_acc])
    plt.plot(train_history.history[test_acc])
    plt.title('Train History')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
show_train_history('accuracy','val_accuracy')
```

```python
show_train_history('loss','val_loss')
```

## 評估模型準確率

```python
scores = model.evaluate(x_Test4D_normalize , y_TestOneHot)
scores[1]
```

- 采用更深的架构(InceptionResNetV2)
- 采用预训练权重进行迁移学习(Pretrained on 'imagenet')
- 进行数据增强(scale, transform, flip .etc)

```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "0,1"
!nvidia-smi

# Helper libraries
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
from os import listdir
from os.path import join
import cv2
import pandas as pd
import random

# step 1: 讀取資料集, 並且 resize 圖片

# Set the path of the input folder 
# data_path = "../input/ml100-03-final/image_data/train/"
data_path = "./image_data/train/"

size = 299, 299
# size = 32, 32

def readImage(path):
    # Import the images and resize them to a 128*128 size
    # Also generate the corresponding labels
    image_names = []
    train_labels = []
    train_images = []

    folders = os.listdir(path)

    for folder in folders:
        for file in os.listdir(os.path.join(path,folder)):
            if file.endswith("jpg"):
                image_names.append(os.path.join(path,folder,file))
                train_labels.append(folder)
                img = cv2.imread(os.path.join(path,folder,file))
                im = cv2.resize(img,size)
                train_images.append(im)
            else:
                continue

    # Transform the image array to a numpy type
    return np.array(train_images), np.array(train_labels), np.array(image_names)
    
train, labels, names = readImage(data_path)

# step 2: labels 轉換成數值, Extract the labels
print(pd.unique(labels))

# 紀錄文字型 labels
txt_label = pd.DataFrame({'txt': labels})

# 文字型 labels 轉成 one_hot_encode
label_dummies = pd.get_dummies(labels)

# 轉成 數值型 labels
labels = label_dummies.values.argmax(1)

# 紀錄數值型 labels
num_label = pd.DataFrame({'val': labels})

# 看一下結果, dataset review
print("train.shape", train.shape)
print("labels.shape", labels.shape)
print("names.shape", names.shape)

import matplotlib.pyplot as plt
def plot_image(image):
    fig = plt.gcf()
    fig.set_size_inches(2, 2)
    plt.imshow(image, cmap='binary')
    plt.show()

plot_image(train[0])
print("labels[0]", labels[0])
print("names[0]", names[0])

# print mapping table
pd_mapping_table = pd.concat([txt_label, num_label], axis=1).drop_duplicates()
mapping_table = pd_mapping_table.set_index('val')['txt'].to_dict()

print(mapping_table)
print(mapping_table[3])
```

```python
# step 3: 切train/test, fix random seed for reproducibility
from sklearn.model_selection import train_test_split
seed = 7
x_train, x_test, y_train, y_test = train_test_split(train, labels, test_size=0.33, random_state=seed)

# step 4: 影像輸入的維度
input_shape = x_train.shape[1:]

# step 5: 資料標準化, 先把影像縮放到 0-1 之間
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# step 6: 對 label 做 one-hot encoding
num_classes = 5
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print("x_train", x_train.shape)
print("x_test", x_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

### model 搭建

```python
import math
import sys

import os, cv2, random
import numpy as np

# 画图工具
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from keras.layers import Input, Dense, GlobalAveragePooling2D
from keras.optimizers import RMSprop
from keras.callbacks import Callback, EarlyStopping, TensorBoard
from keras.applications.inception_resnet_v2 import InceptionResNetV2
from keras.models import Model
from keras.utils import np_utils

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from keras import backend as K
K.set_image_data_format('channels_last') # 数据格式data_format设置为 NHWC
```

### Build DataGenerator
构造训练数据、测试数据、验证数据生成器，使数据在训练、验证和测试过程中分批输入网络。


同时对于训练数据进行了数据增强。

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.inception_v3 import preprocess_input

WIDTH=299
HEIGHT=299
CHANNELS=3
BATCH_SIZE=16
EPOCHS=5
NUM_CLASSES=5

# 訓練集
print("\nTrain Data Set")
train_flow = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=45, 
    width_shift_range=0.2, 
    height_shift_range=0.2, 
    zoom_range=0.3, 
    vertical_flip=True,
    horizontal_flip=True, 
    fill_mode="nearest")


# 將資料送進 ImageDataGenrator 中做增強
train_flow.fit(x_train)

# 測試集
print("\nValidation Data Set")
val_flow = ImageDataGenerator(preprocessing_function=preprocess_input)
# 將資料送進 ImageDataGenrator 中做增強
val_flow.fit(x_test)

#Test DataSet Generator with Augmentation
# print("\nTest Data Set")
# test_generator = ImageDataGenerator(preprocessing_function=preprocess_input)
# test_flow = test_generator.flow_from_directory(
#     test_folder,
#     target_size=(HEIGHT, WIDTH),
#     batch_size = BATCH_SIZE
# )
```

## Model Construct

模型为基于InceptionResNetV2的预训练模型，去除top layer（分类层），在这一基础上冻结预训练参数，并在此基础上增加一个512 hidden unit的全连接层和一个2 hidden unit的分类层。

```python
base_model = InceptionResNetV2(include_top=False, weights='imagenet', input_shape=(WIDTH, HEIGHT, CHANNELS))
```

```python
def build_model(base_model):
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    predictions = Dense(NUM_CLASSES, activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    print(type(model))
    return model
```

```python
for layer in base_model.layers:
    layer.trainable = False

model = build_model(base_model)
model.summary()
```

```python
import keras 
opt = keras.optimizers.rmsprop(lr=0.0001, decay=1e-6)
model.compile(loss='binary_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
es = EarlyStopping(monitor='val_acc', baseline=0.9, patience=15, verbose=1, mode='auto')
cp = ModelCheckpoint(filepath="./ckpt_InceptionResNetV2_flower.h5", monitor="val_acc", verbose=1, save_best_only=True, mode="auto", period=1)
lr = ReduceLROnPlateau(monitor="val_acc", factor=0.1, patience=3, verbose=1, mode="auto", min_lr=0)
callbacks = [es,cp,lr]
```

```python
# 訓練模型囉！
train_history=model.fit_generator(train_flow.flow(x_train, y_train, batch_size=BATCH_SIZE),
                    steps_per_epoch=int(len(x_train)//BATCH_SIZE),
                    validation_data=(x_test, y_test),
                    epochs=EPOCHS, verbose=1,
                    callbacks=callbacks)
# train_history = model.fit_generator(train_flow, 
#                               epochs=5, 
#                               verbose=1,
#                               validation_data=val_flow,
#                               validation_steps=int(len(x_train)//BATCH_SIZE),
#                               steps_per_epoch=int(len(x_test)//BATCH_SIZE),
#                             callbacks=callbacks)

# 評估我們的模型
scores = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
import matplotlib.pyplot as plt
def show_train_history(train,val):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[val])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend([train, val], loc='upper left')
    plt.show()
    
show_train_history('accuracy','val_accuracy')

show_train_history('loss','val_loss')
```

```python
# Helper libraries
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import os
from os import listdir
from os.path import join
import cv2
import pandas as pd
import random

# step 1: 讀取資料集, 並且 resize 圖片

# Set the path of the input folder 
# data_path = "../input/ml100-03-final/image_data/test/"
data_path = "./image_data/test/"

size = 224, 224
# size = 32, 32

def readImage(path):
    # Import the images and resize them to a 128*128 size
    # Also generate the corresponding labels
    image_names = []
    train_labels = []
    val_images = []

    folders = os.listdir(path)

    for file in os.listdir(os.path.join(path)):
            if file.endswith("jpg"):
                image_names.append(os.path.splitext(file)[0])
                img = cv2.imread(os.path.join(path,file))
                im = cv2.resize(img,size)
                val_images.append(im)
            else:
                continue
        

    # Transform the image array to a numpy type
    return np.array(val_images), np.array(image_names)
    
val, names = readImage(data_path)

# normalization

val = val.astype('float32') / 255

# 看一下結果, dataset review
print("train.shape", val.shape)
print("names.shape", names.shape)

import matplotlib.pyplot as plt
def plot_image(image):
    fig = plt.gcf()
    fig.set_size_inches(2, 2)
    plt.imshow(image, cmap='binary')
    plt.show()

plot_image(val[0])
print("names[0]", names[0])
```

```python
# load model and weights
from keras.models import load_model

model_path = './keras_flower_best_trained_model.h5'
model = load_model(model_path)

# 進去預測要做過 normalize
prediction = model.predict(val)
prediction = np.argmax(prediction, axis=1)

output_path = './flower_submission.csv'
pd_output_result = pd.DataFrame({'id': names, 'flower_class': prediction})
pd_output_result.to_csv(output_path, encoding='utf-8', index=False)
pd_output_result.head()
```
